##  <font color= darkred> Data Prep for Modeling </font>

### Train-Test Split

```python
# data prep
from sklearn.model_selection import train_test_split
```

```python
# splitting the data 
X = df.drop(columns = 'booking_status')
y = df['booking_status']
```

```python
# # doing train test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
print("x_train: ", x_train.shape)
print("y_train: ", y_train.shape)


print("\n")

print("x_test: ", x_test.shape)
print("y_test: ", y_test.shape)

display(x_train.head())
display(y_train.head())
```

### Pre-processing 

```python
numerical_cols = x_train.select_dtypes(["int", "float"]).drop(columns = ["year", "month", "day"]).columns
numerical_cols
# x_train[numerical_cols]
```

```python
categorical_cols = x_train.select_dtypes(["object"]).drop(columns = ["month_name", "day_name"]).columns.tolist()

categorical_cols
# x_train[categorical_cols]
```

**Transformation**

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown = "ignore")
```

**We will use ColumnTransformer for handling multiple column types**

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols), # Apply scaler to numerical columns
        ('cat', categorical_transformer, categorical_cols) # Apply one-hot encoding to categorical columns
    ]
)
```

```python
preprocessor
```

## <font color= darkred> Model building </font> üí•‚ö°

```python
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import classification_report
```

```python
pipeline = Pipeline(steps = 
                    [("preprocessor", preprocessor),
                     ("classifier", LogisticRegression())
    
])
```

```python
pipeline
```

```python
# fit the model 
pipeline.fit(x_train, y_train)

# scoring on train 
print(f"Train data accuracy: {pipeline.score(x_train, y_train):.2f}")

# scoring on test 
print(f"Test data accuracy: {pipeline.score(x_test, y_test):.2f}") # conformation on the transformation
```

```python
pipeline
```

## Model Evaluation

**Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# extract the classifier
pipeline.named_steps["classifier"]
```

```python
# extract the classes
lr_class = pipeline.named_steps["classifier"].classes_
lr_class
```

```python
test_predictions = pipeline.predict(x_test)

# generate the confusion matrix
cm = confusion_matrix(y_test, test_predictions, labels = lr_class)
cm = confusion_matrix(y_test, test_predictions)

# display 
disp = ConfusionMatrixDisplay(confusion_matrix = cm,display_labels = lr_class)

# show the plot
disp.plot()
plt.show()
```

**Confusion Matrix Evaluation**

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, test_predictions, target_names= ["Non Cancelled", "Cancelled"]))
```

#### Explanation

**Non-Cancellations**
* Precision: 82% of the items predicted as non-cancellations are actually so (4812/5941)
* Recall: 88% of non-cancelled rooms were correctly classified (4812/5459)

**Cancellations**

* 73% of cancellations predicted are actual cancellations (1792/2439)
* 61% of cancellations were correctly classified (1792/2921); 
    * 2921 --> actual cancellations

### Logistic Regression Model Coefficients

```python
import pprint #https://docs.python.org/3/library/pprint.html
```

```python
feature_names = pipeline.named_steps["preprocessor"].get_feature_names_out()
# print(feature_names,"\n")

# clean the feature names
clean_feature_names = [name.split('__')[-1] for name in feature_names]
# pprint.pp(clean_feature_names)
```

```python
coefficients = pipeline.named_steps["classifier"].coef_[0] # the [0] is to extract the values in single array

# converting into df 
lr_results = pd.DataFrame({"columns": clean_feature_names, "values" : coefficients}).sort_values(by = "values")

# bar plot
lr_results.plot(kind = 'barh', x = 'columns', y = 'values', legend = False);

# Adding labels and title to the plot
plt.title('Logistic Regression Coefficients')

# Display the plot
plt.show()
```

```python
lr_results
```

**Interpretation**
* In logistic regression, the coefficients represent the change in the log-odds of the target variable (cancellation, in this case) 
* The negative value means it decreases the log odds of cancellation. So being a repeated guest has a strong effect on reducing the likelihood of cancellation
* The positive values means it increases the log odds of cancellation.The lead time has a strong effect on increasing the likelihood of cancellation

### ROC - AUC Curve

```python
from sklearn.metrics import roc_curve, auc
```

**Dummy Classifier**

```python
preprocessor
```

```python
# call dummy 
dummy_clf = DummyClassifier()  

# fit dummy 
dummy_clf.fit(preprocessor.fit_transform(x_train) , y_train)

# predict probability
y_pred_prob_dummy = dummy_clf.predict_proba(preprocessor.transform(x_test) )[:, 1]

# compute values
fpr_dummy, tpr_dummy, thresholds_dummy = roc_curve(y_test, y_pred_prob_dummy)

# auc
roc_auc_dummy = auc(fpr_dummy, tpr_dummy)
```

**Logistic Regression**

```python
# get the probability score
y_pred_prob = pipeline.predict_proba(x_test)[:, 1] # get the probability of class 1

# compute values
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

# calculate auc
roc_auc = auc(fpr, tpr)
```

```python
# Plot ROC curve
plt.figure(figsize=(8, 6))

plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot(fpr_dummy, tpr_dummy, color='blue', lw=2, linestyle='--', label='Dummy Classifier (area = %0.2f)' % roc_auc_dummy)

plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')  # Diagonal line (random classifier)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic Curve')
plt.legend(loc='lower right')
plt.show()
```

* A higher AUC value means better performance. A model with AUC close to 1 is very good, while a model with AUC close to 0.5 is performing no better than random guessing.

* ROC: The closer the curve is to the top-left corner, the better the model.
