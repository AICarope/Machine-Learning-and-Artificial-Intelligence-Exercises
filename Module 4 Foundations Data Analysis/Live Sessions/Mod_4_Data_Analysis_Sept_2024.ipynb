# Joining data

```python
# Example DataFrames

# DataFrame 1 (left)
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40]
})

# DataFrame 2 (right)
df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'City': ['New York', 'Chicago', 'San Francisco', 'Seattle'],
    'Salary': [70000, 80000, 90000, 100000]
})
```

```python
df1
```

```python
df2
```

### Inner Join
Inner join returns only rows with matching keys in both DataFrames. The key here is ID.

```python
inner_join = pd.merge(df1, df2, on='ID', how='inner')
inner_join
```

### Left Join
Left join returns all rows from the left DataFrame, and if no match is found in the right DataFrame, NaN is filled in for the missing columns.

```python
left_join = pd.merge(df1, df2, on='ID', how='left')
left_join
```

### Right Join
Right join is the opposite of left join, returning all rows from the right DataFrame and filling missing values from the left DataFrame with NaN.

```python
right_join = pd.merge(df1, df2, on='ID', how='right')
right_join
```

### Outer Join
Outer join returns all rows from both DataFrames and fills NaN where there is no match.

```python
outer_join = pd.merge(df1, df2, on='ID', how='outer')
outer_join
```

### Cross Join
Cross joins (or Cartesian products) combine every row of one DataFrame with every row of another. In pandas, cross joins can be done using merge() with the how='cross' argument. This creates a "Cartesian product" of both tables.

```python
# DataFrame 1
df1 = pd.DataFrame({
    'Color': ['Red', 'Blue']
})

# DataFrame 2
df2 = pd.DataFrame({
    'Size': ['Small', 'Large']
})
```

```python
df1
```

```python
df2
```

```python
cross_join = pd.merge(df1, df2, how='cross')
cross_join
```

### Horizontal Join (Concatenation along columns)
When you want to join two DataFrames side by side (horizontally), you use the pd.concat() function with axis=1. This works similarly to a SQL JOIN but allows you to simply add columns without needing a key.



```python
# DataFrame 1
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

# DataFrame 2
df2 = pd.DataFrame({
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
})
```

```python
df1
```

```python
df2
```

```python
horizontal_concat = pd.concat([df1, df2], axis=1)
horizontal_concat
```

### Vertical Join (Concatenation along rows)
When you want to join two DataFrames by stacking them on top of each other (vertically), you use the pd.concat() function with axis=0. This is used when you have the same columns in both DataFrames and want to append new rows from one to the other.

```python
# DataFrame 3
df3 = pd.DataFrame({
    'ID': [4, 5],
    'Name': ['David', 'Eve'],
    'Age': [40, 45],
    'City': ['San Francisco', 'Seattle']
})
```

```python
df3
```

```python
vertical_concat = pd.concat([horizontal_concat, df3], axis=0)
vertical_concat
```

## Pivoting/Unpivoting
Though not exactly combining, pivoting and unpivoting data is another way to reshape data:

- `pivot()` transforms or reshapes data into a wide format.
- `melt()` transforms wide-format data into a long format, similar to unpivoting in SQL.

### Pivoting

```python
# Data to pivot
df = pd.DataFrame({
    'ID': [1, 2, 1, 2],
    'Year': [2020, 2020, 2021, 2021],
    'Value': [100, 150, 200, 250]
})
```

```python
df
```

```python
pivoted_df = df.pivot(index='ID', columns='Year', values='Value')
pivoted_df
```

### Melting (unpivoting)

```python
# Data to melt
df = pd.DataFrame({
    'Product': ['A', 'B', 'C'],
    'Jan_Sales': [100, 150, 200],
    'Feb_Sales': [110, 160, 210],
    'Mar_Sales': [120, 170, 220]
})
```

```python
df
```

```python
melted_df = pd.melt(df, id_vars=['Product'], var_name='Month', value_name='Sales')
melted_df
```

### Data dictionary

| Variable  | Definition                            | Key                                   |
|-----------|----------------------------------------|---------------------------------------|
| survival  | Survival                              | 0 = No, 1 = Yes                      |
| pclass    | Ticket class                           | 1 = 1st, 2 = 2nd, 3 = 3rd            |
| sex       | Sex                                   |                                       |
| Age       | Age in years                          |                                       |
| sibsp     | # of siblings / spouses aboard Titanic |                                       |
| parch     | # of parents / children aboard Titanic |                                       |
| ticket    | Ticket number                         |                                       |
| fare      | Passenger fare                        |                                       |
| cabin     | Cabin number                          |                                       |
| embarked  | Port of Embarkation                   | C = Cherbourg, Q = Queenstown, S = Southampton |


Variable Notes
- pclass: A proxy for socio-economic status (SES)
  - 1st = Upper
  - 2nd = Middle
  - 3rd = Lower
- age: Age is fractional if less than 1. If the age is estimated, is it in the form of xx.5
- sibsp: The dataset defines family relations in this way...
  - Sibling = brother, sister, stepbrother, stepsister
  - Spouse = husband, wife (mistresses and fiancs were ignored)
- parch: The dataset defines family relations in this way...
  - Parent = mother, father
  - Child = daughter, son, stepdaughter, stepson
  - Some children travelled only with a nanny, therefore parch=0 for them.

```python
# Select only numerical values to create the heatmap
df_numeric = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(10, 8))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```

```python
# Exploring our target variable (0=Didn't survive, 1=Survived)
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df)
plt.title('Survival distribution')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
custom_palette = {'male': 'dodgerblue', 'female': 'mediumvioletred'}

plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=df, palette=custom_palette)
plt.title('Supervival by gender')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
survival_rate_by_gender
```

```python
# Create age bins to group passengers into age ranges
bins = [0, 12, 18, 35, 60, 80]
labels = ['Child', 'Teenager', 'Young Adult', 'Adult', 'Senior']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)

# Calculate survival rate per gender and age group
survival_rate = df.groupby(['Sex', 'age_group'], observed=True)['Survived'].mean().reset_index()

# Plot the survival rate using a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='age_group', y='Survived', hue='Sex', data=survival_rate, palette=custom_palette)

# Add titles and labels
plt.title('Survival Rate by Gender and Age Group', fontsize=16)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.legend(title='Gender')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=80, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
