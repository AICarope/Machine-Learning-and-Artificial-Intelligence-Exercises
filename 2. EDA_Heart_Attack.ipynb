```python
#S T A T I S T I C S (EDA) in df5
#HISTOGRAM GRAPHICAL REPRESENTATION
#Histogram is a freqeuncy distribution chart. Visualize the frequency counts using bar plots to understand the distribution of categories.
#Plot a Histogram for X:
# Histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(heart_attack_1['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

#Analysis: The dataset includes ages ranging from the youngest at 29 years to the oldest at 77 years. The average age of the participants is approximately 54 years, indicating a middle-aged population. The distribution is fairly spread out, with a standard deviation of around 9 years, suggesting some variability in the ages of the participants.
#A histogram of the age data reveals a somewhat normal distribution, with a peak around the mid-50s. This suggests that most participants are in their 50s, which is a critical age range for heart disease risk assessment. Understanding this age distribution is crucial for tailoring heart disease prevention and treatment strategies to the most affected age groups.
```

```python
# Count occurrences of heart attack (output = 1) by sex.This line filters the dataset for rows where a heart attack occurred (output == 1) and counts the occurrences by sex.
heart_attack_by_age = heart_attack_1[heart_attack_1['output'] == 1]['age'].value_counts()
print(heart_attack_by_age)
```

```python
# Count occurrences of heart attack (output = 1) by age
heart_attack_by_age = heart_attack_1[heart_attack_1['output'] == 1]['age'].value_counts()
print(heart_attack_by_age)
```

```python
# Bar plot for Heart Attacks by Age
plt.figure(figsize=(12, 6))
sns.barplot(x=heart_attack_by_age.index, y=heart_attack_by_age.values, palette='viridis')
plt.title('Heart Attack Occurrences by Age')
plt.xlabel('Age')
plt.ylabel('Number of Heart Attacks')
plt.xticks(rotation=45)
plt.show()
#Analysis: The bar graph shows that the ages with the highest number of heart attacks is for:
    #41 to 46 years old
    #51 to 59 years old
    #62 to 67 years old
```

```python
# Bar plot for Heart Attack Occurrences by Sex
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', hue='output', data=heart_attack_1) # the putput =1 filter the participants that have the heart attack.
plt.title('Heart Attack Occurrences by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Heart Attack', labels=['No', 'Yes'])
plt.show()

#Analysis: #Analysis: From the bar graph, a total of 93 males had a heart attack, compare to 72 females. Making it a total of 165 out of 303 participants.
```
